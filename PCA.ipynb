```python
import pandas as pd 
```

```python
df = pd.read_csv("/media/ege/DATA/bamboo/Concat_DFs/20220426_CONCAT_DF.csv")
```

```python
df.fillna(0,inplace=True)
```

```python
df = df.T
df.columns = df.iloc[0]
df = df.reset_index(drop=True)
df = df.drop(0,axis=0)
df = df.drop(1,axis=0)
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
from sklearn.decomposition import PCA

pcs = PCA(n_components=2)
pcs.fit(df)
loadings = pd.DataFrame(pcs.components_, columns=df.columns)
```

```python
loadings
```

```python
import matplotlib.pyplot as plt
sp_pca = PCA()
sp_pca.fit(df)
explained_variance = pd.DataFrame(sp_pca.explained_variance_)

ax = explained_variance.head(10).plot.bar(legend=False, figsize=(8, 4))
ax.set_xlabel('Component')
plt.tight_layout()
plt.show()
```
