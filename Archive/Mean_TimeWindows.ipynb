### For all data

```python
day_paths = list()
base_path="/home/ege/Desktop/project/new/001/{}"
for root, dirs, files in walk("/home/ege/Desktop/project/new/001/"):
    for dirr in dirs:
        if dirr.startswith('2'):
            day_paths.append(dirr)
```

```python
day_paths
```

```python
path_resample = "/home/ege/Desktop/project/new/Resampled_Data/"
```

## 1 Minute Windows

```python
if not exists("/home/ege/Desktop/project/new/Resampled_Data/"):
        mkdir("/home/ege/Desktop/project/new/Resampled_Data/")
```

```python
for day in day_paths:
    mkdir(path_resample+day+"/")
```

```python
for day in day_paths: 
    print(day)
    for root, dirs, files in walk(base_path.format(day)+"/"):
        for file in files:
            print(base_path.format(day)+"/"+file)
            if "checkpoint" in file:
                print(file + " CHECKPOINT")
                pass
            elif file.endswith('.csv'):
                print(file)
                df = pd.read_csv(base_path.format(day)+"/"+file)
                df['ts']=pd.to_datetime(df['ts'], unit='ns', errors='ignore')
                df.set_index('ts',inplace=True)
                df = df.loc[:, ['ltp',"trade_size","level_qty:ask:0","level_qty:bid:0","level_price:ask:0"]]
                df_resample = df.resample('T').mean()
                df_resample.to_csv(path_resample+day+"/"+file)
```
