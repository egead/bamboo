## VWAP Analysis  

**VWAP**: A popular trading indicator that calculates the average price a security has traded at, taking into account both price and volume. 

The VWAP indicator used to evaluate the fair value of an asset and identify potential buying or selling opportunities. By comparing the current price to the VWAP, we can assess the upward or downward trends in the market.

The VWAP is calculated using the following formula:

$\text{VWAP} = \frac{\sum (Price \cdot Volume)}{\sum Volume}$
